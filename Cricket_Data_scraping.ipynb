# Web Scraping ESPNcricinfo
Cricket Records in ICC Cricket World Cup, 2023/24 - India

### Importing required library

```python
from bs4 import BeautifulSoup
from lxml import etree
import requests
import pandas as pd
```

### Getting url content

```python
URL = "https://www.espncricinfo.com/records/tournament/averages-batting-bowling-by-team/icc-cricket-world-cup-2023-24-15338?team=6"  # Replace with the actual URL
r = requests.get(URL)
```

```python
soup = BeautifulSoup(r.content, "html.parser")
```

### Accessing table body use beautiful soup fucntion

```python
table_body = soup.find('table')
print(table_body)
```

### Scraping Function

```python

# Initialize an empty list to store the extracted data
data = []

# Iterate over each row in the table body
for row in table_body.find_all('tr'):
    # Extract data from each cell in the row and append to the data list
    row_data = [cell.text.strip() for cell in row.find_all(['td', 'th'])]
    data.append(row_data)
```

### Converting data in dataframe

```python

num_columns = len(data[0])
expected_columns = 14
if num_columns != expected_columns:
    raise ValueError(f"Expected {expected_columns} columns but found {num_columns} columns")

# Define column names
columns = ["Player", "Mat", "Inns", "NO", "Runs", "BF", "HS", "Ave", "SR", "100", "50", "0", "4s", "6s"]

# Create DataFrame
df = pd.DataFrame(data[1:], columns=columns)

# Print DataFrame
df.head()
```
